# Logistic Regression Using Sklearn

Logistic regression is basically using regression to find a line seperating positive and negative samples

```python
#Importing Dependencies
import pandas as pd
import numpy as np
from sklearn import linear_model
from matplotlib import pyplot
```

```python
#reading data
data=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00267/data_banknote_authentication.txt', header=None)
```

```python
data
```

```python
#randomly arranging the data
data=data.sample(frac=1).reset_index(drop=True)
```

```python
data.head()
```

```python
#declaring the model
logreg=linear_model.LogisticRegression(n_jobs=-1)
```

```python
#splitting the data into training and final evaluation datasets
X=data.loc[0:1320,[0,1,2,3]]
Y=data.loc[0:1320,4]
Xf=data.loc[1320:1372,[0,1,2,3]]
Yf=data.loc[1320:1372,[4]]
```

```python
Yf
```

```python
from sklearn import model_selection
```

```python
#Dividing into training and cross validation data sets
Xtr,Xte,Ytr,Yte = model_selection.train_test_split(X,Y,test_size=0.3,random_state=4,shuffle=True)
```

```python
#Fitting the model
logreg.fit(Xtr,Ytr)
```

```python
logreg.predict(Xte) == Yte
```

```python
from sklearn import metrics
```

## Evaluating our model

```python
metrics.accuracy_score(Yte,logreg.predict(Xte))
```

```python
metrics.accuracy_score(Yf,logreg.predict(Xf))
```

```python
df= pd.DataFrame(Xte)
df['res'] = logreg.predict(Xte)
```

```python
df
```

```python
#taking into account only the first two columns, scatter plot of the dataset
df.plot.scatter(x=0,y=1,c='res',colormap='viridis')
```
