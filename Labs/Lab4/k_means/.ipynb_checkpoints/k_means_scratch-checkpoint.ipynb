# K-Means

K means is a commonly used clustering algorithm used for unsupervised learning. Unsupervised learning, as opposed to supervised learning, is when you don't have any labels for the datapoints. You just have a cluster of datapoints and we want to split it into classes.
<img src='https://raw.githubusercontent.com/ritchieng/machine-learning-stanford/master/w8_unsupervised_learning/unsupervisedlearning6.png'>
Ref: Andew Ng's course on Intro to ML

Now this is important when you have data and want to make some observations based on clusters. For example, in the above data, based on the height and weight of a person, we determine what sized T-Shirt he wears.

K means has a 2 phase algorithm:

Phase 1: Appoint a centroid per class and classify the remaining points based on nearness

Phase 2: Take the average of all datapoints of a class and assign those to the values of centroids.

Phases 1 and 2 are repeated.

In phase 1, generally, random datapoints are chosen as the centroid. Thus, K-Means algorithm is run several times and the best clustering is chosen

And now, k-means must seperate out 2 classes based on the clustering

```python
#Returns random centroids for the data
def centroidsinit(classes,data):
    import random
    #Initializing the centroids atrandom datapoints. Ideally, this assignment must be done many times.
    centroids=[]
    for i in range(classes):
        centroids.append(data.iloc[random.randint(0,len(data)),0:4].tolist())
    
    return centroids
```

```python
centroids1 = centroidsinit(classes=2,data=data)
```

```python
print(centroids1)
```

```python
#returns the new centroids and the corresponding result(clusters)
def kmeans(data,classes,centroidslist,epochs=100):
    #Iterating step
    centroids = centroidslist
    for _ in range(epochs):  #set the epochs as you wish
    
        #Phase1 
        costs=np.zeros((len(data),classes))  #for 'Distance' from the centroids
        for i in range(classes):
            costs[:,i] = ((data.iloc[:,0:4] - centroids[i])**2).sum(axis = 1) #'Distance from the centroids'
        res = costs.argmin(axis=1)  #Choose the index of the minimum loss
    
        #Classification of points based on centroids is done
    
        #Phase 2
        for i in range(len(res)):
            centroids[res[i]] += data.iloc[i,0:4]  
        for i in range(classes):
            deno = (res == i).sum()
            centroids[i] /= deno  #Reassigning with the average of datapoints of that class
        
        #New centroids made
        return centroids,res
```

```python
centroids1, res1 = kmeans(data,classes=2,centroidslist=centroids1)
```
